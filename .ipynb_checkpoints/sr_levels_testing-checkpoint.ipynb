```python
# imports
import ccxt
from pandas import ExcelWriter
import pandas as pd
import numpy as np
import time
import talib
from plotly import graph_objs as go
from mpl_finance import candlestick2_ohlc
import matplotlib.pyplot as plt
from hurst import compute_Hc
import statsmodels.tsa.stattools as ts
import datetime
import collections
import csv
from dateutil import tz
# Auto-detect zones
from_zone = tz.tzutc()
to_zone = tz.tzlocal()

# custom data download function
from bitmex_data_loader import get_bitmex_data

# possible timeframse
timeframes = {'1m': 60*1000,
              '5m': 5*60*1000,
              '15m': 15*60*1000,
              '30m': 30*60*1000,
              '1h': 60*60*1000,
              '3h': 3*60*60*1000,
              '6h': 6*60*60*1000,
              '12h': 12*60*60*1000,
              '1d': 24*60*60*1000,
              '1w': 7*24*60*60*1000,
              '2w': 14*24*60*60*1000,
              '1M': None}

# initiate bitmex to download the data from
bitmex_public = ccxt.bitmex({})
```

```python

# overall chart parameters
symbol = 'BTC/USD'
timeframe = '1h'
candles = 465

# include the current, not yet finished candle
params = {'partial': True}
# when making more than one API call to download the data sleep inbetween calls to not run into the rate limit
api_sleep = 2

# S/R indicator windows
support_resistance_window = 50
support_resistance_window_overall = int(candles * 0.3)

# number of supports / resistances to be plotted on chart
n_levels = 15
# number of candles to extend the S/R lines to the left
extend_left = 10
# number of candles to extend the S/R lines to the right
extend_right = support_resistance_window + 20

usd_support_distance = 10
```

```python
# load BITMEX data

# if the requested timefram is provided by bitmex out of the box
if timeframe in ['1m', '5m', '1h', '1d']:
    # initiate bitmex time now to avoid issues if data is loaded over multiple candles
    bitmex_time_now = bitmex_public.milliseconds()
    if candles > 500:
        runs = candles // 500
        remainder = candles % 500
        data_raw = []
        for i in reversed(range(runs + 1)):
            if i > 0:
                since = bitmex_time_now - (i * 500 * timeframes[timeframe] + remainder * timeframes[timeframe])
                bitmex_data = bitmex_public.fetch_ohlcv(symbol=symbol, timeframe=timeframe, limit=500, since=since, params=params)
                data_raw += bitmex_data
            time.sleep(api_sleep)
        if remainder > 0:
            since = bitmex_time_now - (remainder * timeframes[timeframe])
            bitmex_data = bitmex_public.fetch_ohlcv(symbol=symbol, timeframe=timeframe, limit=remainder, since=since, params=params)
            data_raw += bitmex_data

    else:
        since = bitmex_time_now - candles * timeframes[timeframe]
        data_raw = bitmex_public.fetch_ohlcv(symbol=symbol, timeframe=timeframe, limit=candles, since=since, params=params)

    data_raw = pd.DataFrame.from_records(data_raw)
    data_raw.columns = ['Timestamp', 'Open', 'High', 'Low', 'Close', 'Volume']
    
# if the requested timefram is NOT provided by bitmex out of the box use the custom function
else:
    data_raw = get_bitmex_data(symbol, timeframe, candles, include_current_candle=True, file_format=[], api_cooldown_seconds=api_sleep)

data = data_raw

# get information about timespan of the backtest period
ts_from = data_raw.loc[0, 'Timestamp']
time_from = datetime.datetime.fromtimestamp(ts_from / 1000).strftime('%d.%m.%Y %H:%M:%S')

ts_to = data_raw.loc[len(data_raw) - 1, 'Timestamp']
time_to = datetime.datetime.fromtimestamp(ts_to / 1000).strftime('%d.%m.%Y %H:%M:%S')

total_seconds = float((ts_to - ts_from) / 1000)
day = total_seconds // (24 * 3600)
total_seconds = total_seconds % (24 * 3600)
hour = total_seconds // 3600
total_seconds %= 3600
minutes = total_seconds // 60
total_seconds %= 60
seconds = total_seconds
```

```python
# S/R indicator

def supres(low, high, min_touches=1, stat_likeness_percent=1.5, bounce_percent=5):
    # Setting default values for support and resistance to None
    sup = None
    res = None
    # Identifying local high and local low
    maxima = high.max()
    minima = low.min()
    # Calculating distance between max and min (total price movement)
    move_range = maxima - minima
    # Calculating bounce distance and allowable margin of error for likeness
    move_allowance = move_range * (stat_likeness_percent / 100)
    bounce_distance = move_range * (bounce_percent / 100)
    # Test resistance by iterating through data to check for touches delimited by bounces
    touchdown = 0
    awaiting_bounce = False
    for x in range(0, len(high)):
        if abs(maxima - high[x]) < move_allowance and not awaiting_bounce:
            touchdown = touchdown + 1
            awaiting_bounce = True
        elif abs(maxima - high[x]) > bounce_distance:
            awaiting_bounce = False
    if touchdown >= min_touches:
        res = maxima
        # Test support by iterating through data to check for touches delimited by bounces
        touchdown = 0
        awaiting_bounce = False
    for x in range(0, len(low)):
        if abs(low[x] - minima) < move_allowance and not awaiting_bounce:
            touchdown = touchdown + 1
            awaiting_bounce = True
        elif abs(low[x] - minima) > bounce_distance:
            awaiting_bounce = False
    if touchdown >= min_touches:
        sup = minima
    return sup,res

# dicts where S/R levels were found with keys being the candle index of the S/R level and values being the price
support_d = {}
resistance_d = {}
# fill the above lists and dicts using the above defined function
for i in range(0, len(data) - (support_resistance_window + 1)):
    low = data['Low'][i:i + support_resistance_window].reset_index()
    high = data['High'][i:i + support_resistance_window].reset_index()
    sup, res = supres(low['Low'], high['High'])
    if sup != None:
        support_d[i] = sup
    if res != None:
        resistance_d[i] = res

# construct two dicts with keys being the S/R prices and values being lists holding the candles where the corresponding S/R level is present
support_lines = {}
old_v = 0
for k,v in support_d.items():
    if old_v != v:
        support_lines[v] = [k]
        old_v = v
    else:
        support_lines[v].append(k)
sorted_support_lines = sorted(support_lines.items(), key=lambda kv: max(kv[1]))

resistance_lines = {}
old_v = 0
for k,v in resistance_d.items():
    if old_v != v:
        resistance_lines[v] = [k]
        old_v = v
    else:
        resistance_lines[v].append(k)
sorted_resistance_lines = sorted(resistance_lines.items(), key=lambda kv: max(kv[1]))
```

```python
# candlestick plot

print(time_from + ' to ' + time_to)

fig = plt.figure(figsize=(16,9))
ax = fig.add_subplot(1,1,1)
candlestick2_ohlc(ax, data['Open'], data['High'], data['Low'], data['Close'], width = 0.5, colorup = 'lightgreen', colordown = 'lightcoral', alpha = 0.7)

# plot the minor (on shorter window) S/R levels
if len(sorted_support_lines) >= n_levels:
    for i in range(1, n_levels + 1):
        if len(sorted_support_lines) == 1:
            ax.plot([sorted_support_lines[-i][1][0] - extend_left, sorted_support_lines[-i][1][0] + extend_right], [sorted_support_lines[-i][0], sorted_support_lines[-i][0]], ':', color='blue', label='S ' + str(sorted_support_lines[-i][0]))
        else:
            ax.plot([sorted_support_lines[-i][1][0] - extend_left, sorted_support_lines[-i][1][-1] + extend_right], [sorted_support_lines[-i][0], sorted_support_lines[-i][0]], ':', color='blue', label='S ' + str(sorted_support_lines[-i][0]))
else:
    for i in range(1, len(sorted_support_lines) + 1):
        if len(sorted_support_lines) == 1:
            ax.plot([sorted_support_lines[-i][1][0] - extend_left, sorted_support_lines[-i][1][0] + extend_right], [sorted_support_lines[-i][0], sorted_support_lines[-i][0]], ':', color='blue', label='S ' + str(sorted_support_lines[-i][0]))
        else:
            ax.plot([sorted_support_lines[-i][1][0] - extend_left, sorted_support_lines[-i][1][-1] + extend_right], [sorted_support_lines[-i][0], sorted_support_lines[-i][0]], ':', color='blue', label='S ' + str(sorted_support_lines[-i][0]))
            
if len(sorted_resistance_lines) >= n_levels:
    for i in range(1, n_levels + 1):
        if len(sorted_resistance_lines[-i]) == 1:
            ax.plot([sorted_resistance_lines[-i][1][0] - extend_left, sorted_resistance_lines[-i][1][0] + extend_right], [sorted_resistance_lines[-i][0], sorted_resistance_lines[-i][0]], ':', color='black', label='R ' + str(sorted_resistance_lines[-i][0]))
        else:
            ax.plot([sorted_resistance_lines[-i][1][0] - extend_left, sorted_resistance_lines[-i][1][-1] + extend_right], [sorted_resistance_lines[-i][0], sorted_resistance_lines[-i][0]], ':', color='black', label='R ' + str(sorted_resistance_lines[-i][0]))
else:
    for i in range(1, len(sorted_resistance_lines) + 1):
        if len(sorted_resistance_lines[-i]) == 1:
            ax.plot([sorted_resistance_lines[-i][1][0] - extend_left, sorted_resistance_lines[-i][1][0] + extend_right], [sorted_resistance_lines[-i][0], sorted_resistance_lines[-i][0]], ':', color='black', label='R ' + str(sorted_resistance_lines[-i][0]))
        else:
            ax.plot([sorted_resistance_lines[-i][1][0] - extend_left, sorted_resistance_lines[-i][1][-1] + extend_right], [sorted_resistance_lines[-i][0], sorted_resistance_lines[-i][0]], ':', color='black', label='R ' + str(sorted_resistance_lines[-i][0]))

ax.set_ylabel('Price')
ax.legend(loc='best')
plt.title(str(n_levels) + ' most recent Support / Resistance levels, Timeframe: ' + str(timeframe), y=1.02, fontsize=15 ,fontweight='bold')
ax.grid()
plt.show()
```

```python
# S/R indicator again but for longer chart window

def supres(low, high, min_touches=2, stat_likeness_percent=1.5, bounce_percent=5):
    # Setting default values for support and resistance to None
    sup = None
    res = None
    # Identifying local high and local low
    maxima = high.max()
    minima = low.min()
    # Calculating distance between max and min (total price movement)
    move_range = maxima - minima
    # Calculating bounce distance and allowable margin of error for likeness
    move_allowance = move_range * (stat_likeness_percent / 100)
    bounce_distance = move_range * (bounce_percent / 100)
    # Test resistance by iterating through data to check for touches delimited by bounces
    touchdown = 0
    awaiting_bounce = False
    for x in range(0, len(high)):
        if abs(maxima - high[x]) < move_allowance and not awaiting_bounce:
            touchdown = touchdown + 1
            awaiting_bounce = True
        elif abs(maxima - high[x]) > bounce_distance:
            awaiting_bounce = False
    if touchdown >= min_touches:
        res = maxima
        # Test support by iterating through data to check for touches delimited by bounces
        touchdown = 0
        awaiting_bounce = False
    for x in range(0, len(low)):
        if abs(low[x] - minima) < move_allowance and not awaiting_bounce:
            touchdown = touchdown + 1
            awaiting_bounce = True
        elif abs(low[x] - minima) > bounce_distance:
            awaiting_bounce = False
    if touchdown >= min_touches:
        sup = minima
    return sup,res

# lists containing the prices where S/R levels were found on the longer window
support = []
resistance = []
for i in range(0, len(data) - (support_resistance_window_overall + 1)):
    low = data['Low'][i:i + support_resistance_window_overall].reset_index()
    high = data['High'][i:i + support_resistance_window_overall].reset_index()
    sup, res = supres(low['Low'], high['High'])
    support.append(sup)
    resistance.append(res)

# get unique S/R levels and clean out None values
support = list(set(support))
support = [el for el in support if el != None]
resistance = list(set(resistance))
resistance = [el for el in resistance if el != None]

# sort the S/R levels
support.sort()
resistance.sort()
```

```python
# candlestick plot for S/R levels on longer chart window

print(time_from + ' to ' + time_to)

fig = plt.figure(figsize=(16,9))
ax = fig.add_subplot(1,1,1)

ax.tick_params(axis='both', which='major', labelsize=13)
ax.tick_params(axis='both', which='minor', labelsize=13)

candlestick2_ohlc(ax, data['Open'], data['High'], data['Low'], data['Close'], width = 0.5, colorup = 'lightgreen', colordown = 'lightcoral', alpha = 0.7)

# plot n lowest/highest S/R levels
for n in range(0, n_levels):
    if len(support) >= n_levels:
        ax.axhline(support[n], color='blue')
    else:
        for sup in support:
            ax.axhline(sup, color='blue')
    if len(resistance) >= n_levels:
        ax.axhline(resistance[-n-1], color='black')
    else:
        for res in resistance:
            ax.axhline(res, color='black')

ts_mid = data_raw.loc[int((len(data_raw) - 1) / 2), 'Timestamp']
time_mid = datetime.datetime.fromtimestamp(ts_mid / 1000).strftime('%d.%m.%Y %H:%M:%S')

ax.set_xticks([0, int(len(data) / 2), len(data)])
ax.set_xticklabels([time_from, time_mid, time_to])

ax.set_ylabel('Price', fontsize=20, labelpad=25)
ax.set_xlabel('Date', fontsize=20, labelpad=25)
#ax.legend(loc='best')
plt.title(str(n_levels) + ' (max) lowest Support / highest Resistance levels, Timeframe: ' + str(timeframe), y=1.05, fontsize=20, fontweight='bold')
ax.grid()
plt.show()
```
